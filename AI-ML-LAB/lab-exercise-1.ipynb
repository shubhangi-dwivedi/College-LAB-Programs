# Check
    0-3 - Toddler
    Above 3 - 12 - Kid
    Above 12 - 19 - Teen
    Above 19 - 30 - Young Adults
    Above 30 - 50 - Adults
    Above 50 - 60 - Mature
    Above 60 - 80 - Senior Citizens
    Above 80 - Super Senior Citizens

```python
age = input("Enter age ")
age = int(age)
print(type(age))
if age <=3:
    print("Toddler")
elif age > 3 and age <= 12:
    print("Kid")
elif age > 12 and age <= 19:
    print("Teen")
elif age > 19 and age <= 30:
    print("Young Adult")
elif age > 30 and age < 50:
    print("Adult")
elif age >= 50 and age < 60:
    print("mature")
elif age >= 60 and age < 80:
    print("Senior Citizens")
else:
    print("Super Senior Citizens")
```

```python
#3. WAP to check if roots of a quatratic equation are real, real and equal or imaginary
#quadratic equation : a*x^2 + b*x +c=0

a=input("Enter value of coeff. a :")
b=input("Enter value of coeff. b :")
c=input("Enter value of coeff. c :")

#b^2 -4ac
d= (int(b)*int(b)) - (4*int(a)*int(c))

if(d>0):
    print("Roots for given eq. are REAL")
elif (d==0):
    print("Roots for given eq. are REAL and EQUAL")
else:
    print("Roots are IMAGINARY")
```

```python
"""4. WAP to print grade of a student 
    Less than 40 - NC
    40 - Less than 50 - D
    50 - Less than 60 - C
    60 - Less than 70 - B
    70 - Less than 80 - A
    80 and Above - O"""

y=input("Enter the marks : ")
x=int(y)

print("The grade for given marks is : ")
if(x<40):
    print("NC")
elif(x>=40 and x<50):
    print("D")
elif(x>=50 and x<60):
    print("C")
elif(x>=60 and x<70):
    print("B")
elif(x>=70 and x<80):
    print("A")
else:
    print("O")
```
