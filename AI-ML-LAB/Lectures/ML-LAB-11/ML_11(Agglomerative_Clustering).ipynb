## Aglomerative clustering-hierarchical clustering

```python
import pandas as pd
df=pd.read_csv('Wholesale customers data.csv')
```

```python
df
```

```python
#another technique of preprocessing-normalization
from sklearn.preprocessing import normalize
```

```python
data=normalize(df) #after normaliza it gets converted to array
data=pd.DataFrame(data,columns=df.columns)
data
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
#dendrogram-name of the diagram-give info about no. of clusters
plt.figure(figsize=(10,10))
plt.title("Dendrograms")
dendo=shc.dendrogram(shc.linkage(data,method='ward')) #link between 2 datapoints
#no of clusters-min no of time the line crosses the dendrogram
plt.axhline(y=6,color='r',linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')#linkage-beacuse minimization of the variance
```

```python
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data['Milk'],data['Grocery'],c=cluster.labels_)
```
