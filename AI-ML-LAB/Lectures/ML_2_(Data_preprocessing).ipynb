```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

#for fitting the value of X to convert it in such a way that i'll get the proper distribution
#fitting the data
#and, printing StandardScaler
print(scaler.fit(X)) #fitting X in StandardScaler method to do something, what it is doing is tranformation which is given below i.e. it is only doing scaling
#transforming the data
print(scaler.transform(X))

#we can see that our data lies from -1 to 1 in exponent mentisa form
```

Method for filling missing values

1. Simple imputer

Note:
*   Mean - avg. of all the values
*   Median - middle value of all the data (arranged in ascending or descending order)..when n is odd then (n+1)/2th value, when n is even then (n/2 + (n/2+1))/2th value

->cloumn data is taken for mean or median


```python
from sklearn.impute import SimpleImputer
import numpy as np
#NaN is Not a number

#taking a small data to understand simple imputer
X=[[1,2,3],[4,np.NaN,5],[6,7,8]]
imp=SimpleImputer(missing_values=np.NaN, strategy='mean') #this means wherever np.NaN is found it'll take mean
#it takes means of a column i.e. 2+7/2=4.5

print(imp.fit_transform(X))
```

```python
#we can also use median
X=[[1,2,3],[4,np.NaN,5],[6,7,8]]
imp=SimpleImputer(missing_values=np.NaN, strategy='median') #this means wherever np.NaN is found it'll take median
#it takes median of a column i.e. 2+7/2=4.5

print(imp.fit_transform(X))
```

```python
#we can also use constant
X=[[1,2,3],[4,np.NaN,5],[6,7,8]]
imp=SimpleImputer(missing_values=np.NaN, strategy='constant') #this means wherever np.NaN is found it'll put constant
#whenver NaN will be numeric it'll be replaced by 0

print(imp.fit_transform(X))
```
