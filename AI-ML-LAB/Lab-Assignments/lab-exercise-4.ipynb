# Creating a Matrix

```python
R = int(input("Enter the number of rows:")) 
C = int(input("Enter the number of columns:")) 
  
# Initialize matrix 
matrix = [] 
print("Enter the entries rowwise:") 
  
# For user input 
for i in range(R):          # A for loop for row entries 
    a =[] 
    for j in range(C):      # A for loop for column entries 
         a.append(int(input())) 
    matrix.append(a) 
  
# For printing the matrix 
for i in range(R): 
    for j in range(C): 
        print(matrix[i][j], end = " ") 
    print() 
```

```python
#1.WAP to find minimum element from a list of integers alongwith its index in the list
l=[15,6,9,1,0,8]
min=1000

for i in l:
    if i < min:
        min=i
        
idx=l.index(min)
print("Minimum Element is {}".format(min) , end = "\t")
print("Index is {}".format(idx))
```

```python
#2.WAP to calcualte mean of a given list of numbers
l=[15,6,9,1,0,8]
sum=0
n=len(l)

for i in l:
    sum+=i
avg=sum/n

print("Mean of a given ist of numbers is : ",avg)
```

```python
#3.WAP to search an element in a list
l=[15,6,9,1,0,8]
x=int(input("Enter an element you want to search in the list : "))
y=-1
n=len(l)

flag=0

for i in range(n):
    if x==l[i]:
        flag=1
        y=i
        break

if flag==1:
    print("Element found at position:", y+1)
else:
    print("Element not present in the list")
```

```python
#4.WAP to count the frequency of an element in a list
l=[15,1,6,9,1,0,8]
x=int(input("Enter an element whose frequency you want to count : "))
count=0
n=len(l)

for i in l:
    if x==i:
        count+=1
        
print("Frequency of the element in list is : ",count)
```

```python
#5.WAP to find frequencies of all elements in a list
l=[15,1,8,8,1,0,8,'a','e','i','o','u','a']

#dictionary
freq={}

for i in l:
    if i in freq:
        freq[i]+=1
    else:
        freq[i]=1
        
print('Frequencies of all elements in a list : ',freq)
```

```python
#6. WAP to extact two slices from a list and add the slices and find which one is greater
l=[15,1,8,8,1,0,8]
n=len(l)

mid=n//2
x=l[0:mid]
y=l[mid:n]

sum1=0
sum2=0
for i in x:
    sum1+=i
for i in y:
    sum2+=i
    
print("The list is : ",l)
print("Slice 1 is : ",x)
print("Slice 2 is : ",y)

if sum1>sum2:
    print("Slice 1 is greater")
elif sum2>sum1:
    print("Slice 2 is greater")
else:
    print("Both are equal")
```

```python
#7.WAP to find second largest number of a list
l=[15,6,9,1,0,8]
max1=-999
max2=-1000
for i in l:
    if i>max2:
        if i>max1:
            max2=max1
            max1=i
        else:
            max2=i
            
print("Second largest number of a list : ",max2)
        
```

```python
#8. WAP to input a list of numbers and shift all zeros to the right and all non zeros to left of the list
l=[15,0,6,9,1,0,8]
l1=[]
l2=[]

for i in range (len(l)):
    if l[i]==0:
        l2.append(l[i])
    else:
        l1.append(l[i])
        
l=l1+l2
print("The updated list is : ",l)
```

```python
# 9. WAP to add two matrices
R1 = int(input("Enter the number of rows in matrix :")) 
C1 = int(input("Enter the number of columns in matrix :")) 
  
# Initialize matrix 
matrix1 = [] 
print("Enter the entries rowwise for matrix1:") 
  
# For user input 
for i in range(R1):          # A for loop for row entries 
    a =[] 
    for j in range(C1):      # A for loop for column entries 
         a.append(int(input())) 
    matrix1.append(a) 


# Initialize matrix 
matrix2 = [] 
print("Enter the entries rowwise for matrix2:") 
  
# For user input 
for i in range(R1):          # A for loop for row entries 
    a =[] 
    for j in range(C1):      # A for loop for column entries 
         a.append(int(input())) 
    matrix2.append(a)
    
#adding
for i in range(R1):
    for j in range(C1):
        matrix1[i][j]=matrix1[i][j]+matrix2[i][j]
        

# For printing the matrix 
print("Resultant matrix is : ")
for i in range(R1): 
    for j in range(C1): 
        print(matrix1[i][j], end = " ") 
    print() 
```
