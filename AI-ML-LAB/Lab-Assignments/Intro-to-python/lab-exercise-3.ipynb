```python
#1.WAP to check if a string is Palindrome or not
str=input("Enter a string : ")
n=len(str)

i=0
j=n-1
flag=1

while i<j:
    if str[i]!=str[j]:
        flag=0
        break
    i+=1
    j-=1
        
if flag==0:
    print("Given string is not a palindrome")
else:
    print("Given string is a palindrome")
```

```python
#2.WAP that takes string as multiple words and then capitalize the first letter of each word and forms a new string
str=input("Enter a string : ")
lstr = str.split()
n=len(lstr)

count=0
for i in range(0,n):
    x=lstr[i]
    y=x[0].capitalize()
    lstr[i]=y+x[1:]
    
op_str=""
for i in range(0,n):
    op_str=op_str+lstr[i]+" "
    
print("Result string is : ",op_str)
```

```python
#4.WAP to read a string and then prints a sting that capitalizes every other letter in the string e.g. goel becomes gOeL
str=input("Enter a string : ")
length = len(str)
op_str=""

for i in range(0,length):
    x=""
    if i%2==1:
        x+=str[i].upper()
    else:
        x+=str[i]
    op_str+=x
    x=""
    
print("The output string is :",op_str)
```

```python
#5.WAP to read an email id of a person and vertify if it belongs to banasthali.in domain
str=input("Enter a string : ")
str_find="banasthali.in"

if str.find(str_find)>-1:
    print("email-id is of banasthali.in domain")
else:
    print("email-id is NOT of banasthali.in domain")
```

```python
#6.WAP to count the number of vowels and consonants in an input string
str=input("Enter a string : ")
n=len(str)
v_count=0
c_count=0

i=0
str=str.lower()
while i<n:
    if str[i].isalpha():
        if str[i] in ('a','e','i','o','u'):
            v_count+=1
        else:
            c_count+=1
    i+=1

print("No. of vowels : ",v_count)
print("No. of consonants : ",c_count)
```

```python
#7.WAP to read a sentnece and count the articles it has
str=input("Enter a string : ")
str=str.lower()
lstr = str.split()
n=len(lstr)

count=0
for i in range(0,n):
    if(lstr[i] in('the','an','a')):
        count+=1
        
print("No. of atricles in a sentence : ",count)
```

```python
#8.WAP to check if a sentence is a paniltimate or not
str=input("Enter a string : ")
arr=str.split()
n=len(arr)

if(n>1):
    print("Penultimate word is : ",arr[n-2])
else:
    print("No penultimate word")
```
