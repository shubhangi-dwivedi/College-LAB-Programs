```python
#2.WAP to check if a number is prime or not
import math

n=input("Enter a no. : ")
x=int(n)
flag=0
for i in range(2,int(math.sqrt(x))+1,1):
    if x%i==0:
        flag=1
        break

if flag==1:
    print("Not a prime no.")
else:
    print("It is a prime no.")
    
```

```python
#6.WAP to generate prime series between 1 and 100
import math

print("Prime nos. between 1 and 100 are : ")

flag=0
for i in range(2,100,1):
    for j in range(2,i+1,1):
        if i!=j:
            if i%j==0:
                break
        else:
            print(i)
```

```python
#7.WAP to check if a number is armstrong number or not
n= int(input("Enter a no. : "))
sum = 0

temp = n
while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if n == sum:
    print("Given no. is an Armstrong number")
else:
    print("Given no. is not an Armstrong number")
```

```python
#8.WAP to check if a number is a perfect square or not
import math

n=int(input("Enter a no. : "))
flag=0

for i in range(1,n+1):
    if math.sqrt(n)==i:
        print("It is a perfect square of ", i )
        flag=1
        break
        
if flag==0:
    print("Given no. is not a perfect square")
```

```python
#9.WAP to check if a number is digit palindrome or not
n=int(input("Enter a no. : "))
i=n

new_n=0
while i!=0:
    res=i%10
    new_n=new_n*10+res
    i//=10
    
if n==new_n:
    print("Given no. is a palindrome")
else:
    print("Given no. is not a palindrome")
```

```python
#10.WAP to find factorial of a given number
n=int(input("Enter a no. : "))

fact=1
for i in range (1,n+1):
    fact*=i
print("Factorial of the no. is : ", fact)
```

```python
#11.WAP to check if a year is a leap year or not
n=int(input("Enter a year : "))

if n%400== 0 and n%100 == 0:
    print("Given year is a leap year")
elif n%4==0 and n%100!= 0:
    print("Given year is a leap year")
else:
    print("Given year is not a leap year")
```

```python
#12.WAP to print pascal's triangle
from math import factorial

n= int(input("Enter the number of rows..."))
for i in range(n):
    for j in range(n-i+1):
        print(end=" ")
    for j in range(i+1):
        print(factorial(i)//(factorial(j)*factorial(i-j)), end=" ")
    print()
```

```python
#13.WAP to sum the series 4/3 + 9/4 + 16/5 + 25/6 + ....... nth term
n=int(input("Enter a no. : "))

sq_calc=1
res=0
for i in range(1,n+1):
    sq_calc=(i+1)*(i+1)
    res=res+(sq_calc/(i+2))
print("Sum of series is : ",round(res,3))
```

```python
#14.WAP to find HCM of two numbers
n=int(input("Enter 1st no. : "))
m=int(input("Enter 2nd no. : "))
x=min(n,m)

max_i=1

for i in range (1,x+1):
    if m%i ==0 and n%i==0:
        max_i=max(max_i,i)
        
print("HCM of given nos. is : ",max_i)
```

```python
#15.WAP to calcualte roots of the quadratic equation
import math

a=input("Enter value of coeff. a :")
b=input("Enter value of coeff. b :")
c=input("Enter value of coeff. c :")

#b^2 -4ac
d= (int(b)*int(b)) - (4*int(a)*int(c))

if d>=0:
    x1= (b+ math.sqrt(d))/(2*a)
    x2= (b- math.sqrt(d))/(2*a)
    print("Root x1 :",x1)
    print("Root x2 :",x2)
else:
    print("Roots are imaginary")
```

```python
#16.WAP to generate armstrong numbers between 1 and 500
for i in range(1,500+1):  
    sum = 0  
    temp = i 
    while temp > 0:  
        digit = temp % 10  
        sum += digit ** 3  
        temp //= 10  
        if i == sum:  
            print(i)  
```

```python
#17.WAP to generate perfect squares between 1 and 500
import math

print("Perfect squares between 1 and 500 : ")

for i in range (1,501):
    for j in range (1,i+1):
        if math.sqrt(i)==j:
            print(i)
```

```python
#18.WAP to sum the series 4/2! +9/3! + 16/4! + ..... nth term
n=int(input("Enter a no. : "))

fact=1 
sq_calc=1 
res=0

for i in range(1,n+1): 
    for j in range(1,i+2): 
        fact*=j 
    sq_calc=(i+1)*(i+1) 
    res=res+(sq_calc/fact) 
    fact=1 
print("Sum of series is : ",round(res,3))
```
